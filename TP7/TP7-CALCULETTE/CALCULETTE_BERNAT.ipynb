## PARTIE 3 :  Extraire l'interprétation et calculer le résultat 

```python
def extraire_interpretation(arbre):
    tmp= str(arbre[0])
    racine = tmp.split('\n')
    interpretation = racine[0]
    #print(interpretation)
    expression = interpretation[24:len(interpretation) -2]
    #print('expression = ', expression)
    
    operation = expression.split(", ")
    #print('operation =', operation)
    
    texte_operation = ' '.join(operation)
    
    return texte_operation

'''
Ici, on extrait de notre arbre l'interprétation du calcul et on le met sous forme d'une opération "manuscrite"
(ex : '3 + 5')
'''
```

```python
def calculer(operation):
    resultat = eval(operation)
    #print('resultat = ', operation, ' = ', resultat )
    return resultat
```

## PARTIE 4 :  Application au résultat de la reconnaissance (processus réalisé en amont)

```python
def interpretation_par_lot (Nom_Fichier_Resultat_Reco, Nom_Fichier_Resultat_Calcul, Analyseur,Debug=False):
    ### A Completer

    file = open(Nom_Fichier_Resultat_Reco, 'r')
    operations = file.readlines()
    file.close()

    fileCalcul = open(Nom_Fichier_Resultat_Calcul,"w")

    nb_calcul = 0

    for op in operations:
        t = op.split()
        r = Analyseur.parse(t)
        if Debug:
            fileCalcul.write("================================\n"+r+"\n")
        for a in r:
            o = extraire_interpretation(a)
            e = calculer(o)
            fileCalcul.write("-> "+ o + " = "+ str(e)+"\n")
            nb_calcul+=1


    fileCalcul.close()

    return nb_calcul
```

```python
nb_commande = interpretation_par_lot ("RESULTAT_TRANSCRIPTION_V1.txt", "RESULTAT_CALCUL_V1.txt", analyseur_V1)
if nb_commande:
    print("Nombre de commandes exécutées : ", nb_commande)
else: 
    print("Aucun calcul valide")
```
